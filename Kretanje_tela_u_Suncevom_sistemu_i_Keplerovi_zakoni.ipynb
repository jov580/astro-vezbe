```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt

posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun(posmatrac)

sunce_visine = []

merkur_visine = []
jupiter_visine = []
pluton_visine = []
venera_visine = []
uran_visine = []
mars_visine = []
saturn_visine = []
neptun_visine = []

sunce_rek = []

merkur_rek = []
jupiter_rek = []
pluton_rek = []
venera_rek = []
uran_rek = []
mars_rek = []
saturn_rek = []
neptun_rek = []

for datum in date_range('2024/4/26 9:00:00', '2025/4/26 9:00:00'):
    posmatrac.date = datum
    sunce.compute(posmatrac)
    merkur = Mercury(posmatrac)
    venus = Venus(posmatrac)
    mars = Mars(posmatrac)
    jupiter = Jupiter(posmatrac)
    saturn = Saturn(posmatrac)
    uranus = Uranus(posmatrac)
    neptune = Neptune(posmatrac)
    pluto = Pluto(posmatrac)

    sunce_visine.append(sunce.dec * 180 / pi)

    merkur_visine.append(merkur.dec * 180 / pi)
    venera_visine.append(venus.dec * 180 / pi)
    mars_visine.append(mars.dec * 180 / pi)
    jupiter_visine.append(jupiter.dec * 180 / pi)
    saturn_visine.append(saturn.dec * 180 / pi)
    uran_visine.append(uranus.dec * 180 / pi)
    neptun_visine.append(neptune.dec * 180 / pi)
    pluton_visine.append(pluto.dec * 180 / pi)


    sunce_rek.append(sunce.ra * 180 / pi)

    merkur_rek.append(merkur.ra * 180 / pi)
    venera_rek.append(venus.ra * 180 / pi)
    mars_rek.append(mars.ra * 180 / pi)
    jupiter_rek.append(jupiter.ra * 180 / pi)
    saturn_rek.append(saturn.ra * 180 / pi)
    uran_rek.append(uranus.ra * 180 / pi)
    neptun_rek.append(neptune.ra * 180 / pi)
    pluton_rek.append(pluto.ra * 180 / pi)

plt.figure(figsize=(10, 8))

plt.title('Položaji planeta u Beogradu (26.4.2024-26.4.2025)')
plt.xlabel('Rektascenizija')
plt.ylabel('Deklinacija (u stepenima)')

plt.plot(merkur_rek, merkur_visine, label='Merkur')
plt.plot(venera_rek, venera_visine, label='Venera')
plt.plot(mars_rek, mars_visine, label='Mars')
plt.plot(jupiter_rek, jupiter_visine, label='Jupiter')
plt.plot(saturn_rek, saturn_visine, label='Saturn')
plt.plot(uran_rek, uran_visine, label='Uran')
plt.plot(neptun_rek, neptun_visine, label='Neptun')
plt.plot(pluton_rek, pluton_visine, label='Pluton')
plt.plot(sunce_rek, sunce_visine, label='Sunce', color='orange')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
```

```python
lista = {
    "Merkur": Mercury(),
    "Venera": Venus(),
    "Mars": Mars(),
    "Jupiter": Jupiter(),
    "Saturn": Saturn(),
    "Uran": Uranus(),
    "Neptun": Neptune(),
    "Pluton": Pluto(),
    "Halejeva": readdb("1P/Halley,e,162.2239,58.9763,111.9047,17.872265,0.0130447,0.966321,171.4710,02/10.0/2022,2000,g 4.0,6.0")
}
```

```python
def pozicije_proba(bodies, years=100, start_date=None):
    if start_date is None:
        start_date = datetime.now()

    end_date = start_date + timedelta(days=years*365)
    dates = np.arange(start_date, end_date, timedelta(days=1)).astype(datetime)

    positions = {body: [] for body in lista}

    for date in dates:
        for body_name, body in lista.items():
            body.compute(date)
            positions[body_name].append((body.sun_distance, body.hlon))

    return dates, positions

start_date = datetime.now()
dates, positions = pozicije_proba(lista, years=100, start_date=start_date)
```

```python
def pozicije(positions, years):
    plt.figure(figsize=(12, 12))

    for body_name, pos in positions.items():
        distances = [p[0] for p in pos]
        hlons = [p[1] for p in pos]
        x = [d * np.cos(h) for d, h in zip(distances, hlons)]
        y = [d * np.sin(h) for d, h in zip(distances, hlons)]
        plt.plot(x, y, label=body_name)

    plt.scatter([0], [0], color='yellow', s=25, label='Sunce')  # Sunce u centru
    plt.xlabel('Astronomical Units (AU)')
    plt.ylabel('Astronomical Units (AU)')
    plt.legend()
    plt.grid(True)
    plt.show()

pozicije(positions, years=400)
```

```python
G = 6.67430e-11
M_sun = 1.989e30
AU = 1.496e11

poluose= np.array([
    57.9e9,  # Merkur
    108.2e9, # Venera
    149.6e9, # Zemlja
    227.9e9, # Mars
    778.3e9, # Jupiter
    1427e9,  # Saturn
    2871e9,  # Uran
    4497e9   # Neptun
])

mase = np.array([
    0.330e24,  # Merkur
    4.87e24,   # Venera
    5.97e24,   # Zemlja
    0.642e24,  # Mars
    1898e24,   # Jupiter
    568e24,    # Saturn
    86.8e24,   # Uran
    102e24     # Neptun
])

def period_kepler_original(a):
    return np.sqrt((4 * np.pi**2 * a**3) / (G * M_sun))

T_original = period_kepler_original(poluose)

plt.figure(figsize=(10, 6))
plt.loglog(poluose, T_original, 'o-', label='Originalni Keplerov zakon')
plt.xlabel('Velika poluosa (m)')
plt.ylabel('Period (s)')
plt.title('Period revolucije u odnosu na veliku poluosu (log-log skala)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

def period_kepler_corrected(a, m_planet):
    return np.sqrt((4 * np.pi**2 * a**3) / (G * (M_sun + m_planet)))

T_corrected = period_kepler_corrected(poluose, mase)

plt.figure(figsize=(10, 6))
plt.loglog(poluose, T_corrected, 'o-', label='Korigovani Keplerov zakon')
plt.xlabel('Velika poluosa (m)')
plt.ylabel('Period (s)')
plt.title('Period revolucije u odnosu na veliku poluosu (log-log skala)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

difference = T_corrected - T_original

plt.figure(figsize=(10, 6))
plt.loglog(mase, np.abs(difference), 'o-', label='Razlika u periodima')
plt.xlabel('Masa planete (kg)')
plt.ylabel('Razlika u periodima (s)')
plt.title('Razlika u periodima u odnosu na masu planete (log-log skala)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
